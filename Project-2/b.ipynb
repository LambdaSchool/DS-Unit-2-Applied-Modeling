```python
from category_encoders import OneHotEncoder
from sklearn.pipeline import Pipeline

oneHot = OneHotEncoder( use_cat_names=True)

features_transformed = oneHot.fit_transform(recs[features])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val, w_train, w_val = train_test_split(features_transformed, recs[targets].values, recs[weights])
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

lr_model.score(X_val, y_val)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

lr_permuter = PermutationImportance(lr_model, n_iter=10)

lr_permuter.fit(X_val, y_val)
```

```python
eli5.show_weights(lr_permuter, feature_names=list(X_val.columns))
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr_model = RandomForestRegressor(min_samples_leaf=8, n_jobs=-1, verbose=0)

rfr_model.fit(X_train, y_train)

rfr_model.score(X_val, y_val)
```

```python
rfr_permuter = PermutationImportance(rfr_model, n_iter=1)
rfr_permuter.fit(X_val, y_val)
eli5.show_weights(rfr_permuter, feature_names=list(X_val.columns))
```

```python
from pdpbox.pdp import pdp_isolate, pdp_plot

isolated = pdp_isolate(model=rfr_model, dataset=X_val, model_features=list(X_val.columns), feature='TOTSQFT_EN', num_grid_points=50)
# pdp_plot(isolated, feature_name='TOTSQFT_EN', plot_lines=True)
```

```python
import matplotlib.pyplot as pyplot
pyplot.rcParams['figure.facecolor'] = '#002B36'
pyplot.rcParams['axes.facecolor'] = 'black'
pyplot.rcParams['figure.figsize'] = (10,8)
# pyplot.rcParams['lines.color'] = 'white'

grid = isolated.feature_grids
pdp = isolated.pdp

pyplot.plot(isolated.ice_lines.T, color='blue', alpha=0.025)
pyplot.plot(grid, pdp, color='white', linewidth=3)

pyplot.xlim(0,4000)
pyplot.ylabel('log_btu')
pyplot.xlabel('TOTSQFT_EN')
pyplot.ylim(10,12)
pyplot.show()
```
