## Clean up data frame

```python
#remove display limits
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#copy data
clean_up = main_df
```

```python
#drop timebased or Unnecessary columns
drop = ["gameid","url","league","split","date","week","game","patchno","playerid","player","team",]

clean_up = main_df.drop(columns=drop)
```

```python
#check
clean_up.head()
```

```python
#check for null values
clean_up.isnull().sum()
```

```python
## Fix nan values

#fill nan values with either 1 or 0 
fill_list =[1,0]
clean_up['fbaron'] = clean_up['fbaron'].fillna(pd.Series(np.random.choice(fill_list, size=len(clean_up.index))))

#fill nan values with the column mean
clean_up["fbarontime"]  = clean_up["fbarontime"].fillna((clean_up["fbarontime"].mean()))

#fill nan values with either 1 or 0 
clean_up["herald"] = clean_up["herald"].fillna(pd.Series(np.random.choice(fill_list,size=len(clean_up.index))))

#drop the column mostly nan values
clean_up = clean_up.drop(columns=["heraldtime"])
```

```python
#check
clean_up.isnull().sum()
```

```python
clean_up.head()
```

```python
clean_up.dtypes
```

```python
#change data type object to int

object_int = ["doubles","triples","quadras","pentas"]

for x in object_int:
    clean_up[x] = clean_up[x].replace(' ', np.NaN)
```

```python
for x in object_int:
    #fill nan values with the column mode
    clean_up[x]  = clean_up[x].fillna(clean_up[x].mode()[0])
```

```python
for x in object_int:
    clean_up[x] = clean_up[x].astype(float)
```

```python
clean_up["dmgshare"] = clean_up["dmgshare"].replace(' ', np.NaN)
clean_up["dmgshare"] = clean_up["dmgshare"].astype(float)
clean_up["dmgshare"]  = clean_up["dmgshare"].fillna((clean_up["dmgshare"].mean()))
```

```python
clean_up["earnedgoldshare"] = clean_up["earnedgoldshare"].replace(' ', np.NaN)
clean_up["earnedgoldshare"] = clean_up["earnedgoldshare"].astype(float)
clean_up["earnedgoldshare"]  = clean_up["earnedgoldshare"].fillna((clean_up["earnedgoldshare"].mean()))
```

```python
#drop to many blanks
clean_up["visiblewardclearrate"].value_counts()
```

```python
#drop to many blanks
clean_up["invisiblewardclearrate"].value_counts()
```

```python
clean_up = clean_up.drop(columns=["visiblewardclearrate","invisiblewardclearrate"])
```

```python
clean_up.isnull().sum()
```

```python
#need to drop towers destoryed causing indirect leakage
#teamtowerkills             0
#opptowerkills

clean_up = clean_up.drop(columns=["opptowerkills","teamtowerkills"])
```

## train, validation, test split

```python
from sklearn.model_selection import train_test_split
```

```python
#train and test split 85% and 15% split
train, test = train_test_split(clean_up, train_size=.85, test_size=.15,
                              stratify=clean_up['result'], random_state=42)

clean_up.shape,train.shape,test.shape
```

```python
#train and validation split. split based of test length
train, val = train_test_split(train, test_size = len(test),
                              stratify=train['result'], random_state=42)

clean_up.shape,train.shape,test.shape,val.shape
```

## random forest classifier

```python
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
```

```python
# assign features and target

target = "result"
features = train.columns.drop(target)

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
y_test = test[target]
```

```python
train.select_dtypes(exclude='number').describe().T.sort_values(by='unique')
```

```python
# make the pipeline
pipeline = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='median'), 
    RandomForestClassifier(n_estimators=10, max_depth=3, min_samples_leaf=1, random_state=42)
)

# Fit on train, score on val
pipeline.fit(X_train, y_train)
print('Validation Accuracy', pipeline.score(X_val, y_val))
```

### Random forest importance

## eli5 permuation

```python
import eli5
from eli5.sklearn import PermutationImportance
```

```python
#eli doesnt work with pipline(workaround)
transformers = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='median')
)

X_train_transformed = transformers.fit_transform(X_train)
X_val_transformed = transformers.transform(X_val)

model = RandomForestClassifier(n_estimators=10, max_depth=3, min_samples_leaf=1, random_state=42)
model.fit(X_train_transformed, y_train)
```

```python
permuter = PermutationImportance(
    model, 
    scoring='accuracy', 
    n_iter=5, 
    random_state=42
)

permuter.fit(X_val_transformed, y_val)
```

```python
feature_names = X_val.columns.tolist()

eli5.show_weights(
    permuter,
    top=None, # show permutation importances for all features
    feature_names=feature_names
)
```

## random forest updated after permutation (feat inportance) 

```python
minimum_importance = 0
mask = permuter.feature_importances_ > minimum_importance
features = X_train.columns[mask]
X_train = X_train[features]
```

```python
print('Shape after removing features:', X_train.shape)
```

```python
X_val = X_val[features]

pipeline = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='median'), 
    RandomForestClassifier(n_estimators=10, max_depth=3, min_samples_leaf=1, random_state=42)
)

# Fit on train, score on val
pipeline.fit(X_train, y_train)
print('Validation Accuracy', pipeline.score(X_val, y_val))
```

```python
from sklearn.metrics import accuracy_score
y_pred = pipeline.predict(X_val)
print('Validation Accuracy', accuracy_score(y_val, y_pred))
```

## xgboost model

```python
from xgboost import XGBClassifier
```

```python
encoder = ce.OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train)
X_val_encoded = encoder.transform(X_val)
```

```python
xgb_model = XGBClassifier(
    n_estimators=500, # <= 1000 trees, depends on early stopping
    max_depth=7,# try deeper trees because of high cardinality categoricals
    learning_rate=0.5, # try higher learning rate
    objective="multi:softmax",
    num_class=n+1,
    n_jobs=-1
)
```

```python
eval_set = [(X_train_encoded, y_train), 
            (X_val_encoded, y_val)]
```

```python
xgb_model.fit(X_train_encoded, y_train,
          eval_set=eval_set, 
          eval_metric='merror',
          early_stopping_rounds=50
          )
```

```python
results = xgb_model.evals_result()
train_error = results['validation_0']['merror']
val_error = results['validation_1']['merror']
epoch = range(1, len(train_error)+1)
plt.plot(epoch, train_error, label='Train')
plt.plot(epoch, val_error, label='Validation')
plt.ylabel('Classification Error')
plt.xlabel('Model Complexity (n_estimators)')
plt.ylim((0.18, 0.22)) # Zoom in
plt.legend();
```
